```python
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes( projection='3d' )
x = [ 0,  1, 2 ]
y = [ 0, -1, 2 ]
z = [ -1, 0, 5 ]
ax.plot(x, y, z, '.r')
ax.plot(x, y, z, '-.b')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
for label in ( ax.get_xticklabels() + ax.get_yticklabels() + ax.get_zticklabels() ):
    label.set_fontsize(6)
```

```python
import numpy as np
fig = plt.figure()
ax = plt.axes( projection='3d' )

X = np.array( [ [0, 0, 0], [1, 1, 1] ] )
Y = np.array( [ [0, 1, 2], [0, 1, 1] ] )
Z = np.array( [ [0, 0, 1], [-1, 1, 0.5] ] )

ax.plot_surface(X, Y, Z)

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
```

```python
def f(x, y):
    
    #return x ** 2 + y ** 2
    
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = plt.axes( projection='3d' )

ax.plot_surface(X, Y, Z, cmap = 'autumn')

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
```
